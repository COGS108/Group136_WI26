## Research Question

Is there a statistically significant price difference between personal care products (specifically razors, deodorant, and shampoo) marketed to women versus those marketed to men, when controlling for product size (price per unit) and category? Furthermore, does the magnitude of this price disparity vary across these different product categories?

## Background and Prior Work

The idea that women pay more than men for similar everyday products is often described as the “pink tax.” This term does not refer to a literal tax, but to the observation that personal care products marketed toward women—such as razors, deodorant, and body wash—frequently appear to have higher prices or smaller quantities than comparable products marketed to men. Consumer-focused reporting has highlighted clear shelf-price and size differences. For example, a field comparison discussed by *Texas Standard* based on **Consumer Reports** testing found that women’s deodorants sometimes contained less product for the same price, and women’s razors cost significantly more than men’s razors with similar cartridge counts, even when shaving performance felt similar in use.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) These findings support the claim that gendered marketing is associated with higher per-unit costs for women in some hygiene categories.

However, other economic research suggests that these price gaps do not necessarily indicate discrimination, but instead reflect **product differentiation**. A commentary published by the **Cato Institute** summarizes research showing that when products are matched more carefully by ingredients and features, the average “pink price gap” nearly disappears.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) In particular, economists analyzing personal care products found that once items were restricted to apples-to-apples comparisons—such as matching shampoo or body wash formulations—the remaining price difference was extremely small. This suggests that women’s products may cost more because they include different scents, moisturizing features, packaging, or other attributes rather than simply because of gender labeling.

Industry-focused analysis also shows that products marketed to men and women are often physically designed differently. A detailed **Consumer Reports** feature on razors explains that women’s razors frequently have curved handles, larger lubrication strips, pivot designs meant for body contours, and moisturizing bars, while men’s razors are engineered for dense facial hair and precision trimming.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) These functional distinctions are presented as justifications for price variation. At the same time, research on gender stereotypes in advertising suggests that visual cues—such as pastel colors, smooth aesthetics, and “luxury” presentation—can shape how consumers perceive value before evaluating performance. A study analyzing razor advertising found that participants often expected women’s razors to be more expensive based on design and color alone, linking marketing imagery to price expectations.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

Together, prior work presents two competing explanations: **(1) observed price differences as evidence of gender-based overcharging**, and **(2) price differences as outcomes of differentiated products and consumer preferences**. While media and consumer advocacy emphasize visible price and size gaps, economic studies stress the importance of controlling for measurable product characteristics. Many prior comparisons rely on simple product pairing rather than systematic statistical controls. Our project builds on this debate by directly testing whether a price difference between men’s and women’s personal care products remains after controlling for **product size** and **product category**, helping clarify whether marketing label alone predicts price.

## Hypothesis



We hypothesize that, after controlling for product size (price per unit) and brand, personal care products marketed toward women will have a statistically significantly higher mean price than identical or near-identical products marketed toward men. We predict this disparity will be most pronounced in the "razors" category, as these products often have high visual differentiation despite identical functional mechanisms, allowing for easier price segmentation by manufacturers.


## Data

The ideal dataset for this project would be a comprehensive, real-time table of current retail prices from multiple major national retailers, such as Target, Walmart, CVS, and Amazon. The dataset should include the following variables:

-   Product Name

-   Brand

-   Price

-   Retailer Name

-   Category (e.g., Razor, Shampoo)

-   Marketing Gender (Men / Women / Unisex)

-   Total Size (e.g., 12 oz, 5 count)

-   Unit Price (price per oz or count)

Additional helpful variables would include product ID (unique identifier), product description, and date of data collection.

We would ideally need thousands of observations to ensure statistical power and adequate market representation across brands, product categories, and retailers. The data would be collected by scraping product pages from retailer websites on the same day, in order to control for time-based price fluctuations. The "who" in this dataset would be consumer products (not people), specifically personal care items that are marketed differently by gender.

The data would be stored in a structured **CSV file** with one product per row and each variable in a separate column. This format would make it easy to calculate unit prices and compare similar products across retailers. The dataset could also be stored in a **relational database**, where product ID links to retailer-specific pricing tables.
